```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.to_sql('cliente', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM cliente '
pd.read_sql(query, engine)
```

```python
query = 'UPDATE cliente SET Rendimento_anual=300000 WHERE ID_Cliente=6840104'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('cliente', engine)
```

```python
query = 'DELETE FROM cliente WHERE ID_Cliente=5008809'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('cliente', engine)
```

```python
query = 'INSERT INTO cliente (ID_Cliente, \
    Idade, Grau_escolaridade, Estado_civil, \
    Tamanho_familia, Categoria_de_renda, Ocupacao, \
    Anos_empregado, Rendimento_anual, Tem_carro, \
    Moradia) VALUES(6850985, 33, "Doutorado", "Solteiro", \
    1, "Empregado", "TI", 2, 290000, 0, "Casa/apartamento próprio")'

with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('cliente', engine)
```
